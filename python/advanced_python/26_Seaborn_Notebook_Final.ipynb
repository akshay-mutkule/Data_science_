# Day 26 — Introduction to Seaborn

Seaborn is a powerful Python visualization library built on top of **Matplotlib**. It provides a high-level interface for creating attractive and informative statistical graphics.

---

## Getting Started with Seaborn

### What is Seaborn and Why Use It Over Matplotlib?
- Seaborn is built on top of Matplotlib.
- It simplifies common visualization tasks.
- Provides better default styles and colors.
- Includes powerful tools for **statistical visualization**.

````python
# Installing Seaborn (run only once)
# !pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt

sns.__version__
````

### Importing Seaborn
```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Using Built‑in Datasets
Seaborn provides sample datasets for practice using:

```python
sns.load_dataset("tips")
```
Let's load a dataset and preview it.

````python
tips = sns.load_dataset('tips')
tips.head()
````

---
## Types of Plots
Below are commonly used Seaborn plots with examples.

### 1️⃣ Bar Plot — `sns.barplot()`

````python
sns.barplot(data=tips, x='day', y='total_bill')
plt.title('Average Bill per Day')
plt.show()
````

### 2️⃣ Line Plot — `sns.lineplot()`

````python
sns.lineplot(data=tips, x='size', y='total_bill')
plt.title('Line Plot Example')
plt.show()
````

### 3️⃣ Histogram — `sns.histplot()`

````python
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Distribution of Total Bill')
plt.show()
````

### 4️⃣ Scatter Plot — `sns.scatterplot()`

````python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title('Total Bill vs Tip')
plt.show()
````

### 5️⃣ Box Plot — `sns.boxplot()`

````python
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Bill Amount Distribution by Day')
plt.show()
````

### 6️⃣ Pair Plot — `sns.pairplot()`
Used to explore **relationships between multiple variables**.

````python
sns.pairplot(tips[['total_bill','tip','size']])
plt.show()
````

---
## Customization

### Changing Themes and Styles
Seaborn allows easy styling using:

```python
sns.set_theme()
sns.set_style('darkgrid')
```

````python
sns.set_theme(style='darkgrid')
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Styled Scatter Plot')
plt.show()
````

### Adding Titles and Axis Labels

````python
sns.barplot(data=tips, x='smoker', y='tip')
plt.title('Tips vs Smoking Status')
plt.xlabel('Smoker?')
plt.ylabel('Tip Amount')
plt.show()
````

### Adjusting Colors and Palettes
Seaborn includes many color palettes such as **pastel, deep, bright, muted**.


````python
sns.barplot(data=tips, x='day', y='tip', palette='pastel')
plt.title('Tips by Day (Pastel Palette)')
plt.show()
````
