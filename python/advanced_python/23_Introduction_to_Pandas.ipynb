## Creating a Pandas Series

````python
# Series from a Python List
marks = [78, 85, 90, 66]
series_list = pd.Series(marks)
series_list
````

````python
# Series from a NumPy Array
arr = np.array([10, 20, 30, 40])
series_array = pd.Series(arr)
series_array
````

````python
# Series from a Dictionary
student_scores = {'Math': 88, 'Science': 92, 'English': 81}
series_dict = pd.Series(student_scores)
series_dict
````

## Creating a Pandas DataFrame

````python
# DataFrame from list of dictionaries
data = [
    {'Name': 'Amit', 'Age': 23, 'City': 'Pune'},
    {'Name': 'Neha', 'Age': 25, 'City': 'Mumbai'},
    {'Name': 'Rahul', 'Age': 22, 'City': 'Nagpur'}
]
df_list_dict = pd.DataFrame(data)
df_list_dict
````

````python
# DataFrame from dictionary of lists
data2 = {
    'Name': ['Amit', 'Neha', 'Rahul'],
    'Marks': [85, 90, 78],
    'Passed': [True, True, True]
}
df_dict_list = pd.DataFrame(data2)
df_dict_list
````

````python
# DataFrame from NumPy Array
arr = np.array([
    [1, 'Laptop', 55000],
    [2, 'Mouse', 500],
    [3, 'Keyboard', 1500]
])
df_numpy = pd.DataFrame(arr, columns=['ID', 'Product', 'Price'])
df_numpy
````

````python
# Dataframe from List of Lists
data = [
    ['Rahul', 25],
    ['Priya', 22],
    ['Amit', 23]
]
df = pd.DataFrame(data, columns=['Name', 'Age'])
print(df)
````

````python
# Access Column
print(df['Name'])
````

````python
# Add new column
df['Salary'] = [50000, 60000, 55000]
print(df)
````

````python
# Filter Data
print(df[df['Age'] > 22])
````
