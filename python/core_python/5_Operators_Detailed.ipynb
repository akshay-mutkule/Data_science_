## 1. What are Operators?

Operators are **symbols** used to perform operations on variables and values.

In Python, operators are used to:
- Perform calculations
- Compare values
- Make decisions
- Update variables

Operators are **very important** for logic building and Data Science.

## 2. Arithmetic Operators

Arithmetic operators are used to perform **mathematical calculations**.

| Operator | Description |
|---------|-------------|
| + | Addition |
| - | Subtraction |
| * | Multiplication |
| / | Division |
| // | Floor Division |
| % | Modulus |
| ** | Power |

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

### Real-Life Example (Arithmetic Operators)

Calculating total marks and average.

```python
maths = 80
science = 75
english = 70

total = maths + science + english
average = total / 3

print("Total Marks:", total)
print("Average Marks:", average)
```

## 3. Comparison Operators

Comparison operators are used to **compare two values**.

| Operator | Meaning |
|---------|---------|
| < | Less than |
| > | Greater than |
| <= | Less than or equal to |
| >= | Greater than or equal to |
| == | Equal to |
| != | Not equal to |

```python
x = 5
y = 10

print(x < y)
print(x > y)
print(x <= y)
print(x >= y)
print(x == y)
print(x != y)
```

### Real-Life Example (Comparison Operators)

Checking pass or fail.

```python
marks = 65

print(marks >= 40)
```

## 4. Assignment Operators

Assignment operators are used to **assign and update values** in a variable.

| Operator | Example |
|---------|---------|
| = | x = 10 |
| += | x += 5 |
| -= | x -= 3 |
| *= | x *= 2 |
| /= | x /= 4 |

```python
a = 10
print("Initial value:", a)

a += 5
print("After += 5:", a)

a -= 3
print("After -= 3:", a)

a *= 2
print("After *= 2:", a)

a /= 4
print("After /= 4:", a)
```

## 5. Logical Operators

Logical operators are used to **combine conditions**.

| Operator | Description |
|---------|-------------|
| and | True if both conditions are True |
| or | True if at least one condition is True |
| not | Reverses the result |

```python
x = True
y = False

print(x and y)
print(x or y)
print(not x)
```

### Real-Life Example (Logical Operators)

Checking eligibility for exam.

```python
age = 20
attendance = 80

print(age >= 18 and attendance >= 75)
```

## 6. Operator Precedence

Operator precedence decides **which operation is executed first**.

**Order (High → Low):**
1. `()`
2. `**`
3. `*`, `/`, `//`, `%`
4. `+`, `-`
5. Comparison operators
6. Logical operators

```python
result = 10 + 2 * 3
print(result)

result = (10 + 2) * 3
print(result)
```

- Arithmetic operators are used for math
- Comparison & logical operators are used for conditions
- Assignment operators update values
- Operator precedence is very important in expressions
