## Why We Need Flexible Arguments?
Sometimes:
- We don't know how many values user will pass
- Data keeps changing
- Functions should be reusable

That’s where `*args` and `**kwargs` help.


## *args – Variable Length Positional Arguments

`*args` stores values in a **tuple**.


````python
# Example 1: Print all numbers passed
def show_numbers(*args):
    print(args)

show_numbers(1, 2, 3)
show_numbers(10, 20, 30, 40)
````

````python
# Example 2: Sum of numbers using *args
def sum_numbers(*args):
    total = 0
    for num in args:
        total = total + num
    return total

print(sum_numbers(5, 10))
print(sum_numbers(1, 2, 3, 4, 5))
````

````python
# Example 3: Find maximum number using *args
def find_max(*args):
    return max(args)

print(find_max(10, 3, 7, 25, 1))
````

### Interview Note 🧠
*args internally behaves like a **tuple**.


## **kwargs – Variable Length Keyword Arguments

`**kwargs` stores data in **dictionary form**.


````python
# Example 1: Print keyword arguments
def show_details(**kwargs):
    print(kwargs)

show_details(name="Amit", age=22, course="DS")
````

````python
# Example 2: Loop through kwargs
def employee_info(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

employee_info(name="Riya", role="Intern", salary=20000)
````

````python
# Example 3: Access specific value
def get_city(**kwargs):
    print("City:", kwargs.get("city", "Not Provided"))

get_city(name="Rahul", city="Pune")
get_city(name="Rahul")
````

### Interview Note 🧠
**kwargs behaves like a **dictionary**.


## Using *args and **kwargs Together


````python
# Example: Combine args and kwargs
def demo(*args, **kwargs):
    print("Numbers:", args)
    print("Details:", kwargs)

demo(1, 2, 3, name="Sita", course="DS")
````

## Lambda Functions – One Line Functions

Used for **small logic only**.


```A lambda function is a small, anonymous function (function without a name) used for short, one-line operations.```

### Syntax

```lambda arguments: expression```


- No def keyword

- No function name

- Automatically returns the result

````python
# Example 1: Add two numbers
add = lambda a, b: a + b
print(add(10, 20))
````

````python
# Example 2: Check even or odd
is_even = lambda x: x % 2 == 0
print(is_even(10))
print(is_even(7))
````

##### 1️⃣ Lambda with map() – Transform data

````python
nums = [1, 2, 3, 4]
square = list(map(lambda x: x**2, nums))
print(square)
````

##### 2️⃣ Lambda with filter() – Filter data

````python
nums = [10, 15, 20, 25]
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print(even_nums)
````

##### 3️⃣ Lambda with sorted() – Sorting logic

````python
students = [("Rahul", 85), ("Anita", 92), ("Vikas", 78)]
result = sorted(students, key=lambda x: x[1])
print(result)
````

##### 5️⃣ Lambda for conditional logic

````python
max_num = lambda a, b: a if a > b else b
print(max_num(10, 20))
````

### When NOT to use Lambda ❌
- Large logic
- Multiple conditions
- Readability issues
