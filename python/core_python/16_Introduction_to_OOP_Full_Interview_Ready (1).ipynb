## What is OOP? (Simple Language)

OOP is a way of programming where:
- We create **objects** (real-world things)
- Objects are created from **classes** (blueprints)


## Classes and Objects

- **Class** → Blueprint
- **Object** → Real thing created from class

`CLASSES ARE BLUEPRINT FOR OBJECTS`


````python
# syntax :
# Class :
class ClassName:
    pass or block of code
# 
````

````python
# Example 1: Very simple class and object
class Student:
    pass

s1 = Student()
print(s1)
s2 = Student()
print(s2)
````

### Explanation
- `Student` is a class
- `s1/s2` is an object of Student class


### Class with attributes
##### “Attributes are variables that belong to the object.”

````python
# Syntax

class ClassName:
    attribute1 = value1
    attribute2 = value2

obj = ClassName()

print(obj.attribute1)
print(obj.attribute2)
````

````python
# create a student class which displays student name and his age 

class Student:
    # Student - Class name 
    
    name = "Amit"
    # name -> Attribute (variable of class /object )
    age = 21
    # age -> Attribute (variable of class /object )

s1 = Student()
# s1 → OBJECT created from Student class

print(s1.name)
# Accessing attribute using object

print(s1.age)
# Accessing attribute using object
````

## Methods in a Class

Methods are functions inside a class.


````python
# syntax :
class ClassName:
    def __init__(self, para1,para2):
        self.para1 = para1
        self.para2 = para2
        
    def method_name1(self):
        pass
    .
    .
    .
````

````python
# Example 4: Class with method (create a student class and create a method which greets the students and displays o/p hello my name is student_name)
# class with single method
class Student:
    def __init__(self, name):
        self.name = name
        

    def greet(self):
        print("Hello, my name is", self.name)


s1 = Student("Aman")
s1.greet()
````

````python
#class with multiple method
class Student:
    def __init__(self, name,age,colour):
        self.name = name
        self.age=age
        self.colour=colour
        

    def greet(self):
        print("Hello, my name is", self.name)

    def ages(self):
        print("hello my age is",self.age)
        
    def colours(self):
        print("hello I like the colour",self.colour)


s1 = Student("Aman",24,"red")
s1.greet()
s1.ages()
s1.colours()
````

### Polymorphism Using Classes (Important)


````python
# Syntax :

class ClassA:
    def method_name(self):
        print("Behavior A")

class ClassB:
    def method_name(self):
        print("Behavior B")

obj1 = ClassA()
obj1.method_name()

obj2 = ClassB()
obj2.method_name()
````

````python
# Example 2: Polymorphism using class methods
class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")

a = Dog()
b = Cat()

a.sound()
b.sound()
````

### Key Learning
Same method name `sound()`
Different behavior depending on object
