## Why Sets Are Important
A **set** is used when:
- You want **unique values**
- Order does **not matter**

In Data Science, sets help in:
- Removing duplicates
- Comparing datasets
- Feature analysis

## What is a Set?

A **set**:
- Is an unordered collection
- Stores **unique elements only**
- Uses curly braces `{}`


```python
# Q: Create a set of numbers with duplicates.
numbers = {1, 2, 3, 3, 4, 4, 5}
print(numbers)
```

## Important Property: Uniqueness

Duplicates are **automatically removed**.


```python
# Q: Remove duplicate student IDs using set.
student_ids = [101, 102, 103, 101, 104, 102]
unique_ids = set(student_ids)
print(unique_ids)
```

## Creating an Empty Set (Very Important)

❌ `{}` creates a dictionary  
✅ `set()` creates an empty set


```python
# Q: Create an empty set correctly.
empty_set = set()
print(type(empty_set))
```

## Set is Mutable

You can add or remove elements.


```python
# Q: Add elements to a set.
fruits = {"apple", "banana"}
fruits.add("orange")
print(fruits)
```

## update() Method

Adds multiple elements at once.


```python
# Q: Add multiple elements using update().
colors = {"red", "blue"}
colors.update(["green", "yellow"])
print(colors)
```

## remove() vs discard()

- remove() → error if element not present  
- discard() → no error


```python
# Q: Demonstrate remove and discard.
nums = {1, 2, 3}
nums.remove(2)
# nums.remove(5)   # Error
nums.discard(5)
print(nums)
```

## Set Operations (Very Useful)

- Union
- Intersection
- Difference


```python
# Q: Perform basic set operations.
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # Union
print(A & B)   # Intersection
print(A - B)   # Difference
```

## Membership Testing

Very fast compared to lists.


```python
# Q: Check if an element exists in a set.
students = {"Amit", "Riya", "John"}
print("Riya" in students)
```

## Data Science Connection 🔗
