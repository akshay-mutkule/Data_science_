## Why Loops Are Extremely Important
Loops help us **repeat tasks automatically**.

In Data Science, loops are used for:
- Traversing datasets
- Feature engineering
- Data cleaning
- Model training iterations

Every coding interview checks **loop logic**.


## What is a Loop?
A loop allows us to **execute a block of code multiple times** until a condition is met.

Without loops:
- Code becomes repetitive
- Logic becomes unreadable


## The for Loop

A **for loop** is used when:
- Number of iterations is known
- We are iterating over a sequence


# Example 1 — Loop through a string

```python
# Q: Print "Hello" 5 times.
for i in range(5):
    print("Hello")
```

# Example 2 — Loop through a string

```python
for ch in "python":
    print(ch)
```

# Example 3 — Sum all numbers in a list

```python
total = 0
for n in [3, 5, 7]:
    total += n
print(total)
```

## Understanding range()

range(start, stop, step)


```python
# Q: Print numbers from 1 to 10.
for i in range(1, 11):
    print(i)
```

# Example 1 — Count backward

```python
for i in range(5, 0, -1):
    print(i)
```

# Example 2 — Skip numbers using step

```python
for i in range(0, 11, 2):
    print(i)
```

## Looping Through a List
Very common in real projects.


```python
# Q: Print each item in a grocery list.
groceries = ["milk", "bread", "eggs"]
for item in groceries:
    print(item)
```

# Example 1 — Print names

```python
names = ["Asha", "Ravi", "Neha"]
for n in names:
    print("Hello", n)
```

# Example 2 — Largest number

```python
nums = [5, 12, 9, 21]
largest = nums[0]

for n in nums:
    if n > largest:
        largest = n

print("Largest:", largest)
```

## The while Loop

A **while loop** is used when:
- Number of iterations is unknown
- Loop depends on a condition
- Runs until the condition is false 


```python
# Q: Add 1 to x five times.
x = 0
while x < 5:
    x += 1
    print(x)
```

# Example 1 — Counter

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

# Example 2 — Keep asking until correct

```python
password = ""
while password != "python":
    password = input("Enter password: ")

print("Access Granted")
```

## break Statement

Used to **exit the loop immediately**.


```python
# Q: Stop loop when number reaches 5.
for i in range(10):
    if i == 5:
        break
    print(i)
```

## continue Statement

Used to **skip current iteration**.


```python
# Q: Skip printing number 5.
for i in range(10):
    if i == 5:
        continue
    print(i)
```

## pass Statement

Used when syntax requires a statement but no action is needed.


```python
# Q: Use pass as a placeholder.
for i in range(5):
    if i == 3:
        pass
    print(i)
```

## Nested Loops
Loop inside another loop.


```python
# Q: Print multiplication table from 1 to 3.
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=" ")
    print()
```
