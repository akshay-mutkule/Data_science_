##### File Modes in Python (Very Important for Students)

`When we open a file in Python, we must tell HOW we want to use the file.
This is done using file modes.`

```python
open("filename", "mode")
```

##### Important file modes

| Mode   | Meaning          | File Exists?             | What Happens        |
| ------ | ---------------- | ------------------------ | ------------------- |
| `"r"`  | Read             | Must exist               | Reads data          |
| `"w"`  | Write            | Creates new / Overwrites | Deletes old content |
| `"a"`  | Append           | Creates if not exists    | Adds data at end    |
| `"x"`  | Exclusive create | Must NOT exist           | Error if exists     |
| `"r+"` | Read + Write     | Must exist               | No deletion         |
| `"w+"` | Write + Read     | Creates / Overwrites     | Deletes old content |
| `"a+"` | Append + Read    | Creates if not exists    | Adds at end         |


🔹 1️⃣ "r" – Read Mode

📌 Used to read data from a file

```python
f = open("data.txt", "r")
print(f.read())
f.close()
```

Rules:

❌ Error if file does not exist
❌ Cannot write
✅ Can only read
🧠 Example:
Reading marks from a student record file

🔹 2️⃣ "w" – Write Mode

📌 Used to write new data

```python
f = open("data.txt", "w")
f.write("Hello Students")
f.close()
```

```python
Rules:

✅ Creates file if not exists
⚠️ Deletes old data
❌ Cannot read
🧠 Example:
Writing exam results
⚠️ Interview Warning
👉 "w" mode overwrites the file
```

🔹 3️⃣ "a" – Append Mode

📌 Used to add data at the end

```python
f = open("data.txt", "a")
f.write("\nNew Entry")
f.close()
```

```python
Rules:

✅ Keeps old data
✅ Adds new data at end
❌ Cannot read
🧠 Example:
Adding attendance daily
```

🔹 5️⃣ "r+" – Read & Write

📌 Used for both reading and writing

```python
f = open("data.txt", "r+")
print(f.read())
f.write("\nUpdated")
f.close()
```

```python
Rules:

❌ File must exist
❌ Does NOT delete old content
Cursor starts at beginning
```

🔹 6️⃣ "w+" – Write & Read

📌 Read + Write, but deletes old data

```python
f = open("data.txt", "w+")
f.write("Fresh Data")
f.seek(0)
print(f.read())
f.close()
```

```python
Rules:

⚠️ Old content removed
✅ Creates file if not exists
```

🔹 7️⃣ "a+" – Append & Read

📌 Read + append

```python
f = open("data.txt", "a+")
f.write("\nNew Line")
f.seek(0)
print(f.read())
f.close()
```

#### Simple Memory trick 

| Purpose            | Mode   |
| ------------------ | ------ |
| Read               | `"r"`  |
| Write (delete old) | `"w"`  |
| Add at end         | `"a"`  |
| Safe create        | `"x"`  |
| Read + Write       | `"r+"` |
