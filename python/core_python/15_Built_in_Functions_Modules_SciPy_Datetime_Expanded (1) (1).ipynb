## Date & Time Handling (IMPORTANT)

Used heavily in:
- Logs
- Time-series analysis
- Data tracking


```python
# Example 1: Current date and time
from datetime import datetime
now = datetime.now()
print(now)
```

```python
# Example 2: Extracting parts of date
print("Year:", now.year)
print()
print("Month:", now.month)
print()
print("Day:", now.day)
print()
print("Hour:", now.hour)
```

```python
# Example 3: strftime() – Formatting date
formatted1 = now.strftime("%d-%m-%Y")
formatted2 = now.strftime("%Y/%m/%d")
formatted3 = now.strftime("%d %B %Y, %I:%M %p")

print(formatted1)
print()
print(formatted2)
print()
print(formatted3)
```

```python
# Example 4: Custom log message using datetime
log = f"Program executed on {now.strftime('%d-%m-%Y at %H:%M:%S')}"
print(log)
```

## Geopy Module (Real-World Use Case)
Calculates distance between locations.


```python
# Geopy example (optional)
from geopy.distance import geodesic
pune = (18.5204, 73.8567)
mumbai = (19.0760, 72.8777)
print(geodesic(pune, mumbai).km)

print("Geopy used in logistics & maps")
```
