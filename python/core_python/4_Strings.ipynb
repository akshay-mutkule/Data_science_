## 1. What is a String?

A **string** is a sequence of characters used to store text data.

In Python, strings are **immutable**, meaning once created, they cannot be changed.

Strings can be created using:
- Single quotes `' '`
- Double quotes `" "`
- Triple quotes `''' '''` or `""" """` (for multi-line text)

```python
a = 'Python'
b = "Data Science"
c = '''This is
a multi-line
string'''

print(a)
print(b)
print(c)
```

## 2. String Indexing

Each character in a string has a **position number (index)**.

- Indexing starts from **0**
- Negative indexing starts from **-1** (from the end)

```python
text = "Python"

print(text[0])   # P
print(text[1])   # y
print(text[-1])  # n
print(text[-2])  # o
```

## 3. String Slicing

Slicing is used to extract a portion of a string.

**Syntax:**
`string[start : end : step]`

- `start` → starting index (included)
- `end` → ending index (excluded)
- `step` → jump between characters

```python
text = "DataScience"

print(text[0:4])      # Data
print(text[4:])       # Science
print(text[:4])       # Data
print(text[::2])      # DaaSine
print(text[::-1])     # Reverse string
```

## 4. String Concatenation

Concatenation means **joining strings together**.

We use the `+` operator to concatenate strings.

```python
first_name = "Data"
last_name = "Scientist"

full_name = first_name + " " + last_name
print(full_name)
```

## 5. String Repetition

We can repeat strings using the `*` operator.

```python
print("Python " * 3)
```

## 6. Important String Methods

String methods help us **modify, clean, and analyze text data**.

These methods are heavily used in **Data Science & NLP**.

```python
text = "  Python Programming Language  "

print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("Python", "Data"))
print(text.split())
print(len(text))
```

### Commonly Used String Methods

- `upper()` → Convert to uppercase
- `lower()` → Convert to lowercase
- `strip()` → Remove extra spaces
- `replace()` → Replace text
- `split()` → Convert string to list
- `len()` → Find length of string

## 7. Checking Content in Strings

Python provides methods to check the content of a string.

```python
text = "Python123"

print(text.isalpha())   # Only alphabets
print(text.isdigit())   # Only digits
print(text.isalnum())   # Alphabets + digits
```

## 8. String Formatting

String formatting is used to insert variables into strings.

This is very useful for **printing results, reports, and logs**.

```python
name = "Student"
score = 88

print(f"{name} scored {score} marks")
print("{} scored {} marks".format(name, score))
```

## 9. Real-Life Example

Cleaning user input before storing it in a database.

```python
user_input = "  Python Data Science  "

cleaned_input = user_input.strip().lower()
print(cleaned_input)
```
