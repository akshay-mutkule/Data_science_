## 1️⃣ What is EDA and Why Is It Important?

**Exploratory Data Analysis (EDA)** is the process of:
- Understanding the **structure** of data  
- Detecting **patterns and relationships**
- Finding **outliers or anomalies**
- Checking **assumptions before modeling**

👉 EDA prevents bad conclusions and helps make better decisions.
