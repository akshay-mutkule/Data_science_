## What is an Error? What is an Exception?

### Error
- Problem that stops program
- Example: syntax error

### Exception
- Runtime problem
- Can be handled using try-except


```python
# Example: Syntax Error (cannot be handled)
print("Hello"   # missing bracket
```

```python
# Example: Runtime Exception
print(10 / 0)
```

## Handling Exceptions using try-except

Code that may fail → inside try  
Handling logic → inside except


```python
# Example 1: Division by zero
try:
    a = 10
    b = 0
    print(a / b)
except:
    print("Cannot divide by zero")
```

```python
# Example 2: Handling user input error
try:
    num = int(input("Enter a number: "))
    print(num * 2)
except:
    print("Please enter a valid number")
```

## Handling Specific Exceptions (Important)


```python
# Example 3: Specific exception
try:
    x = int("abc")
except ValueError:
    print("ValueError occurred")
```

```python
# Example 4: Multiple except blocks
try:
    lst = [1, 2, 3]
    print(lst[5])
except ValueError:
    print("Value error")
except IndexError:
    print("Index out of range")
```

## finally Block

What is the purpose of finally?

👉 finally always runs, whether:

- an error occurs ❌
- or no error occurs ✅
- It is mainly used for cleanup work, like:
- closing files
- releasing database connections



```python
# Don't teach if you didnt teach the file Modes 
# Example 5: finally usage
try:
    f = open("test.txt", "w")
    f.write("Hello")
except:
    print("Error occurred")
finally:
    f.close()
    print("File closed")
```

## Raising Exceptions using raise

Used when we want to create our own error


```python
# Example 6: raise keyword
age = int(input("Enter age: "))

if age < 18:
    raise Exception("Age must be 18 or above")
else:
    print("Eligible")
```

## Real-World Examples


```python
# Example 7: ATM withdrawal check
balance = 5000
withdraw = int(input("Enter amount: "))

try:
    if withdraw > balance:
        raise Exception("Insufficient balance")
    balance -= withdraw
    print("Remaining balance:", balance)
except Exception as e:
    print(e)
```
