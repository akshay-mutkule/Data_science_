## Understanding Conditions

A **condition** is an expression that returns:
- `True`
- `False`

Python uses comparison operators:

| Operator | Meaning |
|--------|--------|
| > | Greater than |
| < | Less than |
| >= | Greater than or equal |
| <= | Less than or equal |
| == | Equal to |
| != | Not equal |


```python
# Simple condition example
x = 10
print(x > 5)   # True
print(x < 5)   # False
```

## The if Statement

The `if` statement runs code **only when a condition is true**.




Syntax :if condition:
    statement(s)

```python
# I want to check the eligibility of peoples who can vote 
age = 16

if age >= 18:
    print("Eligible to vote")
    
print("Program End")
```

🧠 Observation:
- If condition is false, Python **skips the block**
- Program does NOT stop


## if-else Statement

Used when **exactly one condition must run**.


```python
num = 12

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

## elif Statement (Multiple Conditions)

Python checks conditions **top to bottom**.
Once matched, remaining conditions are skipped.


```python
temperature = 35

if temperature > 40:
    print("Very Hot")
elif temperature > 30:
    print("Hot")
elif temperature > 20:
    print("Normal")
else:
    print("Cold")
```

## Nested if Statements

An `if` inside another `if`.
Used when decisions depend on multiple checks.


```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Wrong Password")
else:
    print("Invalid User")
```

## Real-Life Example 1: Student Grading System

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 40:
    print("Grade D")
else:
    print("Fail")
```

## Real-Life Example 2: Customer Discount System

```python
bill = float(input("Enter bill amount: "))

if bill >= 5000:
    discount = bill * 0.20
elif bill >= 3000:
    discount = bill * 0.10
else:
    discount = 0

final_amount = bill - discount
print("Final Amount:", final_amount)
```

## Logical Operators with Conditions

| Operator | Meaning |
|--------|--------|
| and | Both conditions must be true |
| or | At least one true |
| not | Reverse condition |


```python
age = 22
has_id = True

if age >= 18 and has_id:
    print("Entry Allowed")
else:
    print("Entry Denied")
```
