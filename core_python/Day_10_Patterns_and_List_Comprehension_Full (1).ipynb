# What is nested loop 

- A nested loop is a loop inside another loop.
- The inner loop runs completely every time the outer loop runs once.

```python
# syntax :

for outer_variable in outer_range:
    for inner_variable in inner_range:
        # code to run again and again
```

```python
Outer loop -> runs 3 times
   Inner loop -> runs 4 times each time
```

```python
# example
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

👉 Outer loop controls rows
👉 Inner loop controls columns

## Simple Pattern Printing

Pattern printing mainly depends on:
- Rows
- Columns
- Nested loops


```python
# Q: Print a simple square pattern of stars (5x5).
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
```

## Right Angle Triangle Pattern

```python
# Q: Print a right angle triangle pattern.
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

## Inverted Right Angle Triangle

```python
# Q: Print inverted right angle triangle.
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(5,0,-1):
    print(i)
```

## Pyramid Pattern

```python
# Q: Print a pyramid pattern.
rows = 5
for i in range(rows):
    for space in range(rows - i - 1):
        print(" ", end="")
    for star in range(2 * i + 1):
        print("*", end="")
    print()
```

## Introduction to List Comprehension

List comprehension provides a **shorter syntax** for creating lists.


#### Syntax :
[expression for item in iterable]

```python
# Example 

squares = [x*x for x in range(1, 6)]
print(squares)
```

```python
# Q: Create a list of squares from 1 to 10 (normal way).
squares = []
for i in range(1, 11):
    squares.append(i ** 2)
print(squares)
```

```python
# Q: Create a list of squares from 1 to 10 (list comprehension).
squares = [i**2 for i in range(1, 11)]
print(squares)
```

## List Comprehension with Condition

```python
# Q: Create a list of even numbers from 1 to 20.
evens = [i for i in range(1, 21) if i % 2 == 0]
print(evens)
```

## Nested List Comprehension

```python
# Q: Create a matrix using list comprehension.
matrix = [[j for j in range(3)] for i in range(3)]
print(matrix)
```
