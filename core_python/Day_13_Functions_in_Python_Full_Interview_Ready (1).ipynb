## What is a Function?

A **function** is a block of code that:
- Runs only when called
- Can accept input
- Can return output


```python
# Syntax :
def funct_name():
    #block of code
    print()

To call function:
print(Funct_name)
```

```python
# Q: Define a simple function that prints a message.
def greet():
    print("Hello, welcome to Data Science journey!")

greet()
```

## Function with Parameters

Parameters allow passing data into functions.


```python
# Q: Function that accepts name as parameter.
def greet_user(name):
    print("Hello", name)

greet_user("Rahul")
```

```python
# Q: Function that accepts multiple parameters
def greet_user(name,age,college):
    print("Hello", name,age,college)

greet_user("Rahul",25,"mit")
```

## Function with Return Value

`return` sends output back to the caller.


🔹 print

Purpose: Displays output on the screen (console)

Returns value: ❌ No (returns None)

Use case: For debugging or showing output to the user

Can be used outside/inside functions

```python
def add(a, b):
    print(a + b)

result = add(2, 3)
print(result)
```

🔹 return

Purpose: Sends a value back from a function

Returns value: ✅ Yes

Use case: When you want to reuse the result in code

Ends function execution

```python
def add(a, b):
    return a + b

result = add(2, 3)
print(result)
```

```python
# Q: Function that adds two numbers and returns result.
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

🔹 Key Difference (One-Line)

print shows the value, return gives the value back to the program

## Recursive Functions

A function calling itself is called **recursion**.


```python
# Q: Find factorial using recursion.
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
