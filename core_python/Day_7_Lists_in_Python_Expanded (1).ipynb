## What is a List?

A **list** is a collection of multiple values stored in a single variable.

### Characteristics:
- Ordered
- Mutable (can be changed)
- Allows duplicate values


```python
# Creating a list
numbers = [10, 20, 30, 40, 50]
print(numbers)
```

## List Indexing

Each element in a list has a position called **index**.
Indexing starts from **0**.


```python
fruits = ["apple", "banana", "mango", "orange"]

print(fruits[0])   # first element
print(fruits[2])   # third element
print(fruits[-1])  # last element
```

## List Slicing

Slicing helps us extract a **portion of the list**.
Syntax:
`list[start : end]`


```python
numbers = [1, 2, 3, 4, 5, 6]

print(numbers[1:4])
print(numbers[:3])
print(numbers[3:])
```

## Modifying Lists (Mutability)

Lists are **mutable**, meaning we can change their values.


```python
marks = [60, 70, 80]
marks[1] = 75
print(marks)
```

## List Methods (Very Important)

### append() – Add one item at the end


```python
grocery = ["milk", "bread"]
grocery.append("eggs")
print(grocery)
```

### extend() – Add multiple items

```python
grocery.extend(["butter", "cheese"])
print(grocery)
```

### insert() – Add item at specific index

```python
grocery.insert(1, "fruits")
print(grocery)
```

### remove() – Remove item by value

```python
grocery.remove("bread")
print(grocery)
```

### pop() – Remove item by index

```python
grocery.pop(2)
print(grocery)
```

### clear() – Remove all items

```python
temp_list = [1, 2, 3]
temp_list.clear()
print(temp_list)
```

### index() & count()

```python
nums = [1, 2, 2, 3, 4]
print(nums.index(2))
print(nums.count(2))
```

### sort() & reverse()

```python
values = [5, 1, 4, 2]
values.sort()
print(values)

```

```python
values.reverse()
print(values)
```

## Real-Life Example: Grocery List Management

```python
grocery = []

grocery.append("Rice")
grocery.append("Oil")
grocery.append("Sugar")

print("My Grocery List:", grocery)
```
