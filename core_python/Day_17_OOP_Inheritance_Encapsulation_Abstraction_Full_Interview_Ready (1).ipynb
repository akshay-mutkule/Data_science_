## What is Inheritance?

Inheritance allows a **child class** to use properties and methods of a **parent class**.

👉 Why?
- Code reusability
- Clean structure


````python
# SYNTAX

class ParentClass:
    # parent class properties and methods
    pass  or block of code

class ChildClass(ParentClass):
    # child class inherits from ParentClass
    pass or block of code 
````

````python
# Example 1: Simple Inheritance (Parent → Child)
class Animal:
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

obj = Dog()
obj.eat()
obj.bark()
````

````python
# Example : create a simple students details using inehritance
````

#### Parent class

````python
class Person:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Name:", self.name)
````

#### Chid class 

````python
class Student(Person):
    def __init__(self, name, roll_no):
        self.roll_no = roll_no
        super().__init__(name)  # calling parent class constructor and passes name to it 

    def show_details(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
````

#### Why we caleed superinit 

When a child class defines its own __init__(), Python does NOT automatically call the parent’s __init__().
So we must call it manually using super().

#### Object Creation

````python
s1 = Student("Amit", 101)

s1.show_name()      # inherited method
print()
s1.show_details()   # child class method
````

### Explanation
- `Animal` → Parent class
- `Dog` → Child class
- Dog can access Animal's method


## Types of Inheritance


### 1️⃣ Single Inheritance
One parent → One child


````python
class Parent:
    def show(self):
        print("This is parent")

class Child(Parent):
    pass

c = Child()
c.show()
````

````python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    pass

d = Dog()
d.sound()
````

### 2️⃣ Multilevel Inheritance
Grandparent → Parent → Child


````python
class Grandparent:
    def house(self):
        print("House from grandparent")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

c = Child()
c.house() 
````

````python
class Person:
    def show_name(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("I am a student")

class Intern(Student):
    def work(self):
        print("I am an intern")

i = Intern()
i.show_name()
i.study()
i.work()
````

### 3️⃣ Multiple Inheritance
One child → Multiple parents


````python
class Father:
    def skill1(self):
        print("Driving")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skill1()
c.skill2()
````

````python
class Father:
    def skills(self):
        print("Father: Driving")

class Mother:
    def skills(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skills()   # Father method (MRO)
````

#### One-Line Definitions 

Single Inheritance: One child inherits from one parent

Multilevel Inheritance: A class inherits from another derived class

Multiple Inheritance: A class inherits from more than one parent

## Encapsulation

Encapsulation means **binding data and methods together** and controlling access.
Encapsulation is the OOP concept of wrapping data (variables) and methods together and restricting direct access to data.
It helps in data security, control, and code safety.

👉 Achieved using:
- Public
- Private members


#### Public

- Can be accessed from anywhere
- No underscore

````python
class Student:
    name = "Amit"   # public
````

````python
s = Student()
print(s.name)
````

#### Private

- Cannot be accessed directly
- Uses __ (double underscore)

````python
class Student:
    __age = 21   # private
````

````python
s = Student()
print(s.__age)   # ❌ ERROR
````

✔ Access using method:

````python
def get_age(self):
    return self.__age
````

````python
class Student:
    def __init__(self, age):
        self.__age = age   # private variable

    def get_age(self):
        return self.__age  # accessing private variable INSIDE class
````

````python
s1 = Student(21)
print(s1.get_age())
````

````python
# Syntax:

class ClassName:
    def __init__(self):
        self.public_var = 10        # public
        self.__private_var = 20     # private
````

````python
# Create an ATM machine using encapsulation (keep information private)

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance    # private variable

    def get_balance(self):
        return self.__balance       # controlled access through creating method 

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
````

````python
# create the object 
acc = BankAccount(5000)
````

````python
# print(acc.__balance)   # not allowed as we cant access private variable directly 
print(acc.get_balance()) # ✅ allowed
print()

acc.deposit(1000)
print(acc.get_balance())
````

### Key Point
- `__balance` cannot be accessed directly
- Data is protected


## Abstraction (Introduction)

Abstraction means **showing only essential details** and hiding implementation.
```so we can also say ,Abstraction is the OOP concept of hiding implementation details and showing only essential features to the user.```
👉 User knows *what* to do, not *how* it works.


Real-Life Example (Easy)

🚗 Car

- You use accelerator, brake, steering

- You don’t know how engine works

👉 This is abstraction

````python
# SYNTAX

from abc import ABC, abstractmethod

class ClassName(ABC):

    @abstractmethod
    def method_name(self):
        pass
````

### Simple Abstraction Example


Q.
Write a Python program using abstraction to define an abstract class Shape with an abstract method area().
Implement the method in a derived class Square.

#### Abstract Class

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
````

#### Child Class

````python
class Square(Shape):
    def area(self):
        print("Area = side * side")
````

#### Object Creation

````python
s = Square()
s.area()
````

### Explanation
- Abstract class cannot be used directly
- Child must implement abstract method
