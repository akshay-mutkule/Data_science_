## What is Regex? (Again, Very Simple)

Regex helps us **find patterns** in text.

Think of it as:
- CTRL + F but **smarter**


## Regex Basics You MUST Know

| Pattern | Meaning |
|-------|--------|
| \d | digit (0-9) |
| \w | letter/number |
| \s | space |
| + | one or more |
| * | zero or more |
| {n} | exact count |


## re.search() – finds ONLY THE FIRST MATCH in the string.


```python
import re
import warnings 
warnings.filterwarnings("ignore")

text = "Order ID is 4567"
result = re.search("\d+", text)
print(result.group())
```

#### What does group() mean?
- result is a Match object
- group() returns the exact text that was matched

```python
text = "No number here 123"
result = re.search("\d+", text)
print(result.group())
```

```python
# What is want to match more than 1 group of digits in the same sequnce
text = "my name is prasad 1234 and my roll no is 2456"
result= re.search("\d+",text)
print(result.group())
```

## re.match() – Match Only at Start


```python
text = "123 is the code"
print(re.match("\d+", text).group())
```

```python
text = "Code is 123"
print(re.match("\d+", text))
```

## re.findall() – Find ALL Matches


```python
text = "Prices: 100 200 300"
print(re.findall("\d+", text))
```

```python
text = "Phones: 9876543210, 9123456780"
print(re.findall("\d{10}", text))
```

```python
text = "prasad.charate11@gmail.com"
print(re.findall("\d+",text))
```

## re.sub() – Replace Patterns


```python
# Syntax
re.sub(pattern, replacement, text)
```

```python
text = "My number is 9876543210"
print(re.sub("\d{10}", "XXXXXXXXXX", text))
```

```python
text = "Hello   World"
print(re.sub("\s+", "*", text))
```

## re.split() – Split Using Pattern


```python
text = "apple,banana;grapes orange"
print(re.split("[,; ]", text))
```

```python
text = "2024-05-10"
print(re.split("-", text))
```

## Practical Use Case 1: Extract Emails


```python
#  extract all the names from the emails
text = "Contact: Hi this are all the emails : amitshah@gmail.com, prasad@yahoo.com"
print(re.findall("\w+@\w+\.com", text))
```

```python
#  extract all the names from the emails
text = "Contact: Lokesh , Aniket , Deepali, amitshah@gmail.com, prasad@yahoo.com"
print(re.findall("\w+", text))
```

```python
#  extract all the names whose mail id is present
text = "Contact: Lokesh , Aniket , Deepali, amitshah@gmail.com, prasad@yahoo.com"
print(re.findall("\w+@", text))
```

```python
#  extract all the domain names from the emails
text = "Contact: Hi this are all the emails : amitshah@gmail.com, prasad@yahoo.com"
print(re.findall("@\w+\.com", text))
```

## Practical Use Case 2: Extract Phone Numbers


```python
text = "Call 9876543210 or 9123456780"
print(re.findall("\d{10}", text))
```

## Practical Use Case 3: Subtitle Timestamps


```python
text = "00:01 Start 00:15 Middle 00:45 End"
print(re.findall("\d{2}:\d{2}", text))
```

## Practical Use Case 4: Clean Text Data


###### ^ INSIDE square brackets [ ] → NOT --> Means: match anything EXCEPT a–z

```python
import re
text = "abc123"
print(re.sub("[^a-z]", "", text))
```

- abc --> allowed 
- 123 --> removed
` ^ think this as exclude`

```python
text = "Hello!!! Welcome@@@"
clean = re.sub("[^a-zA-Z ]", "", text)
print(clean)
```

`Simple English explanation
Remove everything except alphabets and spaces.`

##### Keep only digits

```python
re.sub("[^0-9]", "", text)
```

##### Keep alphanumeric

```python
re.sub("[^a-zA-Z0-9 ]", "", text)
```
