## What is a Dictionary?

A **dictionary**:
- Stores data as key–value pairs
- Uses curly braces `{}`
- Keys must be **unique**
- Keys are usually strings or numbers


```python
# Q: Create a dictionary storing student details.
student = {
    "name": "Rahul",
    "age": 21,
    "course": "Data Science"
}
print(student)
```

## Accessing Dictionary Values

Values are accessed using **keys**, not index.


```python
# Q: Access name and course from dictionary.
print(student["name"])
print(student["course"])
```

## Updating Dictionary Values

Dictionaries are **mutable**.


```python
# Q: Update age and add a new key.
student["age"] = 22
student["college"] = "ABC Institute"
print(student)
```

## Removing Items from Dictionary

```python
# Q: Remove a key using pop().
student.pop("college")
print(student)
```

## Dictionary Methods

Commonly used methods:
- keys()
- values()
- items()


```python
# Q: Print keys, values, and items.
print(student.keys())
print(student.values())
print(student.items())
```

## Looping Through a Dictionary

```python
# Q: Loop through dictionary and print key-value pairs.
for key, value in student.items():
    print(key, ":", value)
```

## Nested Dictionaries

Used to store structured data.


```python
# Q: Store details of multiple students.
students = {
    "student1": {"name": "Amit", "age": 20},
    "student2": {"name": "Riya", "age": 21}
}
print(students)
```

## JSON-like Dictionary Structure

APIs return data mostly in JSON format (dictionary-like).


```python
# Q: Example of JSON-style weather data.
weather_data = {
    "city": "Pune",
    "temperature": 30,
    "humidity": 70
}
print(weather_data)
```

## Data Science Connection 🔗
