## Abstraction (Introduction)

Abstraction means **showing only essential details** and hiding implementation.
```so we can also say ,Abstraction is the OOP concept of hiding implementation details and showing only essential features to the user.```
👉 User knows *what* to do, not *how* it works.

Abstraction means:
- Showing **only what is necessary**
- Hiding **internal implementation details**

👉 User focuses on *using* the system, not *building* it.


Real-Life Example (Easy)

🚗 Car

- You use accelerator, brake, steering
- You don’t know how engine works

👉 This is abstraction

````python
# SYNTAX

from abc import ABC, abstractmethod

class ClassName(ABC):

    @abstractmethod
    def method_name(self):
        pass
````

Q.
Write a Python program using abstraction to define an abstract class Shape with an abstract method area().
Implement the method in a derived class Square.

#### Abstract Class

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
````

#### Child Class

````python
class Square(Shape):
    def area(self):
        print("Area = side * side")
````

#### Object Creation

````python
s = Square()
s.area()
````

### Explanation
- Abstract class cannot be used directly
- Child must implement abstract method

## CAR as an Abstraction

````python
from abc import ABC, abstractmethod

# Abstract class
class Car(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass
````

````python
# Concrete class implementing abstraction
class PetrolCar(Car):
    def start(self):
        print("Petrol car starts with key")

    def stop(self):
        print("Petrol car stops using brakes")

car = PetrolCar()
car.start()
car.stop()
````

### Explanation
- User knows `start()` and `stop()`
- Engine logic is hidden


````python
# Another implementation
class ElectricCar(Car):
    def start(self):
        print("Electric car starts with button")

    def stop(self):
        print("Electric car stops automatically")

ecar = ElectricCar()
ecar.start()
ecar.stop()
````

## ATM as an Abstraction


````python
# Abstract ATM class
class ATM(ABC):
    @abstractmethod
    def withdraw(self):
        pass

    @abstractmethod
    def check_balance(self):
        pass
````

````python
# ATM implementation
class SBI_ATM(ATM):
    def withdraw(self):
        print("Cash withdrawn successfully")

    def check_balance(self):
        print("Balance shown on screen")

atm = SBI_ATM()
atm.withdraw()
atm.check_balance()
````

ATM is an abstract class

- It is like a blueprint / rule book
- It does not do any real work

It only says:

- “Any class that claims to be an ATM must have
withdraw() and check_balance() methods”

💡 Important:

- You cannot create an object of ATM
- These methods have no implementation (pass)

Think of it as:

- “Every ATM must be able to withdraw cash and show balance —
I don’t care how, just do it.”

````python
class SBI_ATM(ATM):
    def withdraw(self):
        print("Cash withdrawn successfully")

    def check_balance(self):
        print("Balance shown on screen")


This is called Method Overriding.

🔹 What is happening here?
SBI_ATM inherits from ATM

ATM says:
👉 “You MUST implement withdraw() and check_balance()”

SBI_ATM says:
👉 “Okay, here is how SBI ATM does it”
````

### Key Learning
- User does not see:
  - Server logic
  - Database logic
  - Security checks
