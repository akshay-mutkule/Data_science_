## Python Type Casting

```python
It is used to change the datatype of variable.
We need to specify the variable datatype in which we want to convert it.

int
float
complex

int to float
int to complex
float to int
float to complex
complex to int
complex to float
```

#### 1. int to float

```python
a = 45
print(f"Value of a is = {a}, dataype = {type(a)}")

b = float(a)

print("After Converting ----->")
print(f"Value of a is = {b}, dataype = {type(b)}")
```

```python
a = 45
print(f"Original dataype = {type(a)}")

b = float(a)

print(f"After Converting dataype = {type(b)}")
```

#### 2. int to complex

```python
a = 45  # int
print(f"Original dataype = {type(a)}")

b = complex(a)

print(b)
print(f"After Converting dataype = {type(b)}")
```

#### 3. float to int

```python
num = 10.9999
print(f"Original dataype = {type(num)}")

num2 = int(num)

print(num2)
print(f"After Converting dataype = {type(num2)}")
```

#### 4. float to complex

```python
num = 10.9999
print(f"Original dataype = {type(num)}")

num2 = complex(num)

print(num2)
print(f"After Converting dataype = {type(num2)}")
```
